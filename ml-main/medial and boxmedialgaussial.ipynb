```python
#Applying different types of blur efffect..
#Median blur
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#Median blur
box=cv2.medianBlur(img,1) #we can take any positive odd number. As the number increases, the effect increases.
plt.subplot(1,2,1), plt.imshow(img), plt.title('Source image')
plt.subplot(1,2,2), plt.imshow(box), plt.title('Median blur effect')
```

```python

#Applying different types of blur efffect..
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#Box blur
box=cv2.blur(img, (10,10)) # can take any positive number. As the number increases, the effect increases,
#Goussian blur
gaus=cv2.GaussianBlur (img, (- 12, - 12), 10)# de can take any add number. As the number tecreases, the effect increases. colculates the effect of bur by itself. You can have any number in place of a.
#median blur
med=cv2.medianBlur(img,1)# we can take any positive add number. As the number increases, the effect increases.
plt.figure(figsize=(10,3))
plt.subplot(2,2,1), plt.imshow(img), plt.title('Source inage')
plt.subplot(2,2,2), plt.imshow(box), plt.title('Box blur effect')
plt.subplot(2,2,3), plt.imshow(gaus),plt.title('Gaussian blur effect')
plt.subplot(2,2,4), plt.imshow(med), plt.title('Median blur effect')
```
