```python
#Contrast effect on Gray Scale image
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('au.jpg',0)
contrast=cv2.equalizeHist(img)
plt.subplot(1,2,1), plt.imshow(img, cmap= 'gray'), plt.title('Source'), plt.axis('off')
plt.subplot(1,2,2), plt.imshow(contrast, cmap="gray"), plt.title('Contrasti image using equalizaHist'), plt.axis('off')
```

```python
#Contrast effect on color image
#Using Global Histogram Equalization (GHE)
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('au.jpg')
#Convert to YCrCb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y, cr, cb=cv2.split(ycrcb)
#Global Histogram Equalization
y_eq=cv2.equalizeHist(y)
ycrcb_eq=cv2.merge([y_eq,cr,cb])
global_eq=cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1), plt.imshow(source), plt.title('Source'), plt.axis('off')
plt.subplot(1,2,2), plt.imshow(global_eq), plt.title('Contrast color image using equalizeHist'), plt.axis('off')
```

```python
#contrasteffect on color image
#Using CLAHE
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('au.jpg')
#Convert to YCrCb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
y,cr,cb=cv2.split(ycrcb)
#Global Histogram Equalization
clahe=cv2.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
y_clahe=clahe.apply(y)
ycrcb_clahe=cv2.merge([y_clahe,cr,cb])
clahe_eq=cv2.cvtColor(ycrcb_clahe,cv2.COLOR_YCrCb2RGB)
source=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1), plt.imshow(source), plt.title('Source'),plt.axis('off')
plt.subplot(2,2,2), plt.imshow(clahe_eq), plt.title('Contrast color image using CLAHE'), plt.axis('off')
```

```python
#Object detection from the image
#face detection
import cv2
#idad pre-trained foce detector (Haar Cascade)
face_cascade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
#Aflead image
img=cv2.imread('au.jpg')
gray_image=(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY))
#Detect faces
face=face_cascade.detectMultiScale(gray_image,scaleFactor=1.1,minNeighbors=4)
print(face)
#Oras rectangles around the faces
for(x,y,w,h) in face:
    cv2.rectangle(img, (x,y), (x+w,y+h), (0,255,0),3)
cv2.imshow('Face Detector', img)
cv2.waitKey()
cv2.destroyAllWindows()
```
