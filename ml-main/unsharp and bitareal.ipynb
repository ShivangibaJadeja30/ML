```python

#applying advance image filters
#1). Bilaterol (for advanced smoothing effect)
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('au.jpg')
bilateral=cv2.bilateralFilter(img,d=9,sigmaColor=95, sigmaSpace=95)
plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)), plt.title('Source'), plt.axis('off')
plt.subplot(1,2,2), plt.imshow(cv2.cvtColor(bilateral, cv2.COLOR_BGR2RGB)), plt.title('With smoothing effect'), plt.axis('off')
```

```python
#Applying advance inoge filters
#2). Unsharp (for advanced sharpening effect)
import cv2
import matplotlib.pyplot as pit
#Taking input
img=cv2.imread('au.jpg')
#Blur the Leoge
gauss=cv2.GaussianBlur(img, (9,9),0)
#Unsharp masking
unsharp=cv2.addWeighted(img,1.5,gauss,-0.5,0)
plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Source'), plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)), plt.title('with sharpening effect'), plt.axis('off')
```
