```python
#Edge Detection
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Edge detection works on intensity values, not on color. So Gray laage is netter option.
#Goussion blur
Gaussian=cv2.GaussianBlur(img, (3,3),8)
#Canny (src, thresholdl, threshold2)
canny=cv2.Canny (Gaussian, threshold1=50, threshold2=158)
#aploting the trage
plt.subplot(1,2,1),plt.imshow(img), plt.title('Source Image'), plt.axis('off')
plt.subplot(1,2,2), plt.imshow(canny,cmap='gray'), plt.title('Canny image'),plt.axis('off')
#plt.subplot(1, 3,3), plt.imshow(laplactan bacmap-gray).pit tizing cupluctan Image,pit.arts('off')
```
