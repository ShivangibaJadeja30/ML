```python
#Image Sharpening...
#Gaussian Kernel to generate more natural enhancement.
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gauss=cv2.GaussianBlur(img, (5,5),8)
sharped_img=cv2.addWeighted(img,1.5,gauss,-0.5,0)
plt.subplot(1,2,1), plt.imshow(img),plt.title('Source image'), plt.axis('off')
plt.subplot(1,2,2), plt.imshow(sharped_img), plt.title('sharped_img'),plt.axis('off')
```
