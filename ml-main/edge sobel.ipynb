```python
#Edge Detection
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('au.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #Edge detection works on intensity values, not on color. So Gray image
#Sobel (image, ddepth, x, y, kernel_size)
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1,10)
sobel_combined=cv2.magnitude (sobelx, sobely) #The magnitude combines strong edges, regardless of horizontal or ve
#Ploting the image
plt.figure(figsize=(10,6))
plt.subplot(1,4,1), plt.imshow(img),plt.title('Source Image'),plt.axis("off")
plt.subplot(1,4,2), plt.imshow(sobelx,cmap='gray'), plt.title('Sobelx image'),plt.axis('off')
plt.subplot(1,4,3), plt.imshow(sobely,cmap='gray'), plt.title('Sobely image'), plt.axis('off')
plt.subplot(1,4,4), plt.imshow(sobel_combined,cmap='gray'),plt.title('Sobel_combined image'), plt.axis('off')
```
